Cette section présente une <strong>analyse expérimentale détaillée</strong> de l’algorithme Tabou appliqué à un ensemble de tests répétés sur une même instance du problème.  
L’objectif est de quantifier la stabilité, la robustesse et l’efficacité de la méthode à travers des mesures statistiques, tout en identifiant les marges d’amélioration possibles.
</div>

<h3 style="color:#333; font-size:18px; margin-top:16px;">1. Méthodologie expérimentale</h3>
<div style="
    background-color:#F9FAFB;
    border-left: 5px solid #4A90E2;
    padding: 14px 20px;
    border-radius: 8px;
    margin-top: 10px;
    font-size: 15px;
    line-height: 1.6;
    text-align: justify;
">
L’expérience consiste à exécuter **20 lancements indépendants** de l’algorithme Tabou sur une même matrice de distances.  

Chaque exécution fournit :  
<ul>
  <li>Le coût total de la meilleure solution trouvée</li>
  <li>Le temps d’exécution correspondant</li>
</ul>
Les résultats sont ensuite stockés dans un **DataFrame Pandas**, permettant une analyse statistique complète (moyenne, écart-type, IC95%, Z-scores, etc.).
</div>

<h3 style="color:#333; font-size:18px; margin-top:16px;">2. Analyse statistique globale</h3>
<div style="
    background-color:#F9FAFB;
    border-left: 5px solid #4A90E2;
    padding: 14px 20px;
    border-radius: 8px;
    margin-top: 10px;
    font-size: 15px;
    line-height: 1.6;
    text-align: justify;
">
Les indicateurs suivants sont calculés :  
<ul>
  <li><strong>Moyenne</strong> du coût total (<code>mean_time</code>)</li>
  <li><strong>Médiane</strong> (<code>median_time</code>)</li>
  <li><strong>Écart-type</strong> (<code>std_time</code>) — mesure de dispersion</li>
  <li><strong>Minimum et maximum</strong> des coûts</li>
  <li><strong>Gap</strong> = 100 × ((moyenne - minimum)/minimum)</li>
  <li><strong>Intervalle de confiance à 95 %</strong> (IC95%) pour la moyenne</li>
</ul>
Ces valeurs permettent d’évaluer la **variabilité des résultats** et la **convergence** de l’algorithme.
</div>

<h3 style="color:#333; font-size:18px; margin-top:16px;">3. Visualisation et interprétation des résultats</h3>
<div style="
    background-color:#F9FAFB;
    border-left: 5px solid #4A90E2;
    padding: 14px 20px;
    border-radius: 8px;
    margin-top: 10px;
    font-size: 15px;
    line-height: 1.6;
    text-align: justify;
">
Plusieurs graphiques sont générés avec Matplotlib :  

<b>a) Boxplot enrichi — distribution des coûts totaux</b>  
<ul>
  <li>Montre la dispersion des coûts sur les 20 lancements</li>
  <li>Annotations : moyenne, médiane, min, max, IC95%</li>
  <li>Valeurs extrêmes indiquent les runs les plus performants ou faibles</li>
</ul>

<b>b) Courbe du coût total par run</b>  
<ul>
  <li>Visualise la stabilité de l’algorithme au fil des exécutions</li>
  <li>Ligne horizontale = moyenne, zone grisée = IC95%</li>
</ul>

<b>c) Barres du gap (%)</b>  
<ul>
  <li>Exprime l’écart relatif de chaque run par rapport au meilleur résultat</li>
  <li>Ligne rouge (7%) = seuil de tolérance pour identifier les runs moins performants</li>
</ul>

<b>d) Z-score plot — détection d’anomalies</b>  
<ul>
  <li>Runs avec |Z| > 2 → potentiels outliers (piège local ou mauvais paramétrage)</li>
  <li>Permet de savoir le nombre de runs anormales</li> 
</ul>
</div>

<h3 style="color:#333; font-size:18px; margin-top:16px;">4. Interprétation et discussion</h3>
<div style="
    background-color:#F9FAFB;
    border-left: 5px solid #4A90E2;
    padding: 14px 20px;
    border-radius: 8px;
    margin-top: 10px;
    font-size: 15px;
    line-height: 1.6;
    text-align: justify;
">
Les graphiques et indicateurs montrent généralement une bonne cohérence des résultats :  
<ul>
  <li>L’écart-type reste modéré → stabilité algorithmique</li>
  <li>Quelques runs peuvent avoir un Z-score > 2 → sensibilité aux conditions initiales ou paramètres comme le nombre de camions</li>
  <li>Gap moyen < 10% → bonne performance moyenne</li>
</ul>
L’IC95% offre une estimation fiable du coût moyen attendu, consolidant la validité statistique des observations.

<h3 style="color:#333; font-size:18px; margin-top:16px;">5. Limites et perspectives d’amélioration</h3>
<div style="
    background-color:#F9FAFB;
    border-left: 5px solid #4A90E2;
    padding: 14px 20px;
    border-radius: 8px;
    margin-top: 10px;
    font-size: 15px;
    line-height: 1.6;
    text-align: justify;
">
Les variations observées d’un run à l’autre montrent :  
<ul>
  <li>Dépendance aux conditions initiales</li>
  <li>Possible stagnation prématurée (optimum local)</li>
  <li>Dispersion du temps de calcul liée à la nature heuristique</li>
</ul>
<b>Pistes d’amélioration :</b>
<ul>
  <li>Diversification adaptative pour éviter stagnation</li>
  <li>Adapter dynamiquement la taille de la liste Tabou selon la phase</li>
  <li>Combiner Tabou avec multi-start intelligent ou hybridation avec recuit simulé</li>
</ul>
</div>
